```python
idade = 17
if idade >= 18:
  print('Você pode beber cerveja!')
if idade <= 17:
  print('você não pode beber!')
```

```python
#Como resolve o mesmo problema com otimização de codigo
idade = 18
if idade >= 18:
  print('Você pode beber cerveja!')
else:
  print('você não pode beber!')
```

```python
#Como usar elif
idade = 17
if idade >= 70:
  print('Cuidado com a bebida')
elif idade >= 18:
  print('Você pode beber cerveja!')
elif idade == 17:
  print('você ainda não pode beber! espera mais 1 ano')
else:
  print('você não pode beber!')
```

```python
# 1.Faça um programa que vende uma garrafa de água:
# A - se o cliente escolheu água mineral natural, será cobrado R$1,50
# B - se o cliente escolhe água mineral com gás, será cobrado R$2,50

texto = input('Escolhe a sua água para compra (1) Água mineral Natural ou (2) água mineral com gás')

if texto == '1':
  print('A sua conta deu:R$1,50')
elif texto == '2':
  print('A sua conta deu R$2,50')
else:
  print('Digita novamente')
```
